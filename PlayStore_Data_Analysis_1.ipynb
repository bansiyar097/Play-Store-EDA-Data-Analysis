**bold text**# **Q6. Add a new column Least Android Version in the above dataset which contains leastandroid version. These values come from the Android Ver column numerical values.(Android Ver = (4.0.3 and up) => Least Android Version = (4.0.3))**

```python
df["Android Ver"].unique()
```

```python
df["Least Android Version"] = df["Android Ver"]
```

```python
df["Least Android Version"] = df["Least Android Version"].str.replace("and up","").str.replace("Varies with device",df["Android Ver"].mode().iloc[0])
```

```python
df["Least Android Version"].unique()
```

```python
df["Least Android Version"] = df["Least Android Version"].str.replace("and up","").str.replace("W","")
```

```python
def rohan(val):
  x = val.split(" ")
  if len(x) >1:
    return x[0]
  else:
    return val
```

```python
df["Least Android Version"] = df["Least Android Version"].apply(rohan)
```

```python
df["Least Android Version"].unique()
```
